```python
# import complete dataset and the list of features for each model
comp_features = pd.read_csv(feature_path+"complete_merged_features.csv")

# import list of (current) features for each model
features_static = pd.read_csv(feature_path+"featurelist_static.csv")['0'].to_list()
features_dynamic = pd.read_csv(feature_path+"featurelist_dynamic.csv")['0'].to_list()
```

## 1. First Model - Static

```python
static = comp_features[features_static]
static
```

#### Comorbidity

```python
# Format Comorbidity
comorbidity = pd.read_csv(feature_path+'Comorbidity/complete_24hr_comorbidity.csv')
comorbidity.drop(["Unnamed: 0"], axis=1, inplace=True)
comorbidity = comorbidity.add_prefix("comorb_")
comorbidity.rename(columns={"comorb_patientunitstayid" : "patientunitstayid"}, inplace=True)
comorbidity.columns
```

```python
comorbidity
```

```python
len(comorbidity.patientunitstayid.unique())
```

- processsed 'inner' merge to get more accurate information (it is better to drop the rows of the patients (n=3269) who doesn't have the comorbidity information than imputing 0 value to all comorbidity-related columns)

```python
# merging to the current dataframe
static = static.merge(comorbidity, how="inner", on="patientunitstayid")
static
```

#### Medication_Anticholinergics

```python
# Format Comorbidity
antich = pd.read_csv(feature_path+'Medications/Anticholinergics_beforeAdmission.csv')[['patientunitstayid','Anticholinergics']]
antich
```

```python
# merging to the current dataframe
static = static.merge(antich, how="inner", on="patientunitstayid")
static
```

#### Final dataset

```python
print("# positive case:", static[static['Class']==1].shape[0])
```

```python
# drop unnecessary columns
static.drop(['final_offset','apacheadmissiondx'], axis=1, inplace=True)
```

```python
static.isna().sum()[static.isna().sum()!=0]
```

```python
# print the final columns
for colname in static.columns:
    print(colname)
```

```python
# save off the feature list
feature_list = pd.DataFrame(np.array(static.columns))
feature_list.to_csv(modelling_path+"comp_featurelist_static.csv", index=False)
feature_list
```

```python
# save off the dataframe for the classification
static.to_csv(modelling_path+"prediction_df_static.csv", index=False)
# check if the save was successful
pd.read_csv(modelling_path+"prediction_df_static.csv")
```

## 2. Second Model - Dynamic

```python
dynamic = comp_features[features_dynamic]
dynamic
```

#### Comorbidity

```python
# Format Comorbidity
comorbidity = pd.read_csv(feature_path+'Comorbidity/complete_24hr_comorbidity.csv')
comorbidity.drop(["Unnamed: 0"], axis=1, inplace=True)
comorbidity = comorbidity.add_prefix("comorb_")
comorbidity.rename(columns={"comorb_patientunitstayid" : "patientunitstayid"}, inplace=True)
print(comorbidity.columns)
comorbidity
```

- processsed 'inner' merge to get more accurate information (it is better to drop the rows of the patients (n=3269) who doesn't have the comorbidity information than imputing 0 value to all comorbidity-related columns)

```python
# merging to the current dataframe
dynamic = dynamic.merge(comorbidity, how="inner", on="patientunitstayid")
dynamic
```

#### Medications

```python
# Format medication
med = pd.read_csv(feature_path+'Medications/24hrsAllDrugFeatures.csv')
med.drop(["Unnamed: 0"], axis=1, inplace=True)
med.columns
```

```python
# merging to the current dataframe
dynamic = dynamic.merge(med, how="inner", on="patientunitstayid")
dynamic
```

#### Final dataset

```python
# only the patient who has been diagnosed after 24hr
dynamic = dynamic[dynamic.final_offset>1440]
print(dynamic.shape)
dynamic
```

```python
# drop unnecessary columns
dynamic.drop(['final_offset','apacheadmissiondx'], axis=1, inplace=True)
```

```python
print("# positive case:", dynamic[dynamic['Class']==1].shape[0])
```

```python
# print the final columns
for colname in dynamic.columns:
    print(colname)
```

```python
# save off the feature list
feature_list = pd.DataFrame(np.array(dynamic.columns))
feature_list.to_csv(modelling_path+"comp_featurelist_dynamic.csv", index=False)
feature_list
```

```python
# save off the dataframe for the classification
dynamic.to_csv(modelling_path+"prediction_df_dynamic.csv", index=False)
# check if the save was successful
pd.read_csv(modelling_path+"prediction_df_dynamic.csv")
```
